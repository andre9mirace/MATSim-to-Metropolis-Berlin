```python
def set_size(width=470, ratio="golden", fraction=1.0):
    """Set figure dimensions to avoid scaling in LaTeX.

    Parameters
    ----------
    width: float
            Document textwidth or columnwidth in pts
    fraction: float, optional
            Fraction of the width which you wish the figure to occupy

    Returns
    -------
    fig_dim: tuple
            Dimensions of figure in inches
    """
    # Width of figure (in pts)
    fig_width_pt = width * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27
    if ratio == "golden":
        # Golden ratio to set aesthetic figure height
        # https://disq.us/p/2940ij3
        ratio = (5**0.5 - 1) / 2
    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * float(ratio)
    fig_dim = (fig_width_in, fig_height_in)
    return fig_dim

def get_figure(width=470, ratio="golden", fraction=1.0):
    fig, ax = plt.subplots(figsize=set_size(width, ratio, fraction))
    return fig, ax
```

# Nodes

```python
def read_nodes():

    node_data = []

    for event, elem in ET.iterparse(xopen(NETWORK_PATH, "r"), events=["start"]):
        if elem.tag == "node":
            try:
                node_id = (elem.attrib["id"])
                x = float(elem.attrib["x"])
                y = float(elem.attrib["y"])
                node_data.append((node_id, x, y))
            except Exception as e:
                print(f"Error al procesar nodo: {e}")
                
        if event == "end" and elem.tag == "node":
            elem.clear()
            
    nodes_df = pd.DataFrame(node_data, columns=["node_id", "x", "y"])
    
    # Limpiar solo los IDs que tienen "cluster"
    nodes_df.loc[nodes_df["node_id"].str.contains("cluster"), "node_id"] = (
        nodes_df.loc[nodes_df["node_id"].str.contains("cluster"), "node_id"]
        .str.replace("cluster_", "", regex=False)
        .str.split("_")
        .str[0]
    )
    nodes_df["node_id"] = nodes_df["node_id"].astype(int)
    
    return nodes_df
```

```python
def read_network():
    tree = ET.iterparse(xopen(NETWORK_PATH, "r"), events=["start", "end"])
    links = []
    
    for xml_event, elem in tree:
        if elem.tag == "link" and xml_event == "start":
            atts = elem.attrib
            
            atts["link_id"] = atts["id"].replace("#", "")
            atts["from_node"] = atts.pop("from")
            atts["to_node"] = atts.pop("to")
             
            if "cluster" in atts["from_node"]:
                atts["from_node"] = atts["from_node"].replace("cluster_", "").split("_")[0]
            if "cluster" in atts["to_node"]:
                atts["to_node"] = atts["to_node"].replace("cluster_", "").split("_")[0]
            
            atts["length"] = float(atts["length"])
            atts["freespeed"] = float(atts["freespeed"])
            atts["capacity"] = float(atts["capacity"])
            atts["permlanes"] = float(atts["permlanes"])
            atts["is_generated"] = False
            atts["generated_from"] = None
            
            if "volume" in atts:
                atts["volume"] = float(atts["volume"])
                
            links.append(atts)
            
        if elem.tag in ["node", "link"] and xml_event == "end":
            elem.clear()
    
    links = pd.DataFrame.from_records(links)
    links = links.loc[links["modes"].str.contains("car")].copy()
    links["link_id"] = links["link_id"].astype(int)
    links["from_node"] = links["from_node"].astype(int)
    links["to_node"] = links["to_node"].astype(int)
    
    node_pair_counts = links[["from_node", "to_node"]].value_counts()
    if node_pair_counts.max() > 2:
        print("More than two parallel edges")
        
    parallel_idx = node_pair_counts.loc[node_pair_counts > 1].index
    if len(parallel_idx):
        print("Found {} parallel edges".format(len(parallel_idx)))
        next_node_id = max(links["from_node"].max(), links["to_node"].max()) + 1
        next_link_id = links["link_id"].max() + 1
        new_rows = []

        for (source, target) in parallel_idx:
            mask = (links["from_node"] == source) & (links["to_node"] == target)
            idx = mask[mask].index

            # Generate a new node and two split edges for the second edge
            original_edge = links.loc[idx[0]]
            to_split_edge = links.loc[idx[1]].copy()

            # Update the to_node of the second edge to the new node
            links.loc[idx[1], "to_node"] = next_node_id
            links.loc[idx[1], "is_generated"] = True
            links.loc[idx[1], "generated_from"] = original_edge["link_id"]

            # Create the new short edge from the new node to the original target
            new_edge = to_split_edge.copy()
            new_edge["from_node"] = next_node_id
            new_edge["to_node"] = original_edge["to_node"]
            new_edge["link_id"] = next_link_id
            new_edge["length"] = 0.0
            new_edge["is_generated"] = True
            new_edge["generated_from"] = original_edge["link_id"]
            
            new_rows.append(new_edge)

            next_node_id += 1
            next_link_id += 1

        links = pd.concat((links, pd.DataFrame(new_rows)), ignore_index=True)
        
    # Add node coordenates to all links    
    nodes = read_nodes()

    coord_dict = dict(zip(nodes["node_id"], zip(nodes["x"], nodes["y"])))
    
    # from_x, from_y
    links["from_x"] = links["from_node"].map(lambda node_id: coord_dict.get(node_id, (None, None))[0])
    links["from_y"] = links["from_node"].map(lambda node_id: coord_dict.get(node_id, (None, None))[1])

    # to_x, to_y
    links["to_x"] = links["to_node"].map(lambda node_id: coord_dict.get(node_id, (None, None))[0])
    links["to_y"] = links["to_node"].map(lambda node_id: coord_dict.get(node_id, (None, None))[1])
    
    # Dict link to node for matching
    link_to_node = links.set_index("link_id")["to_node"].to_dict()

    for idx, row in links.iterrows():
        if row["is_generated"] and pd.notna(row["generated_from"]):
            origin_link_id = row["generated_from"]
            origin_to_node = link_to_node.get(origin_link_id)

            if origin_to_node is not None:
                origin_coord = coord_dict.get(origin_to_node)

                if origin_coord is not None:
                    if pd.isna(row["from_x"]) or pd.isna(row["from_y"]):
                        links.at[idx, "from_x"] = origin_coord[0]
                        links.at[idx, "from_y"] = origin_coord[1]

                    if pd.isna(row["to_x"]) or pd.isna(row["to_y"]):
                        links.at[idx, "to_x"] = origin_coord[0]
                        links.at[idx, "to_y"] = origin_coord[1]
                        
    links = links.drop(columns = ["is_generated", "generated_from"])
    
    return links
```

# Graphs

```python
def travel_time_scatter(filename, legs, routes, mp_trips):
    m = 180  # Maximum travel time to display (in minutes).
    mp_trips = (
        mp_trips.filter(~pl.col("length").is_null())
        .with_columns((pl.col("arrival_time") - pl.col("departure_time") 
                      ).alias("travel_time"))
        .sort("trip_id")
    )
    legs = legs.loc[legs["id"].isin(mp_trips["trip_id"])].copy()
    legs.sort_values("id", inplace=True)
    routes = routes.copy()

    # Add a join for legs and routes to add the travel_time to legs
    legs = pd.merge(legs, routes, how="left", on=["id", "id"])
    
    legs["trav_time"] = pd.to_timedelta(legs["trav_time"]).dt.total_seconds()
    fig, ax = get_figure(fraction=0.8)
    ax.scatter(
        legs["trav_time"] / 60, 
        mp_trips["travel_time"] / 60,
        #s=20,
        marker=".",
        alpha=0.002, # Opacit√©
        color=CMP(1)
    )
    ax.plot([0, m], [0, m], color=CMP(2), alpha=0.9)
    ax.set_xlabel("MATSim travel time (min.)")
    ax.set_ylabel("METROPOLIS2 travel time (min.)")
    ax.set_xlim(0, m)
    ax.set_ylim(0, m)
    
    ticks = list(range(0, m+1, 20))  # ticks every 20 minutes
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.grid(True, linestyle="--", alpha=0.5)
    
    fig.tight_layout()
    if filename:
        fig.savefig(filename, dpi=300)
    else:
        fig.show()
```

```python
def travel_time_distribution(filename, legs, routes, mp_trips):
    m = 120  # Maximum travel time to display (in minutes).
    mp_trips = (
        mp_trips.filter(~pl.col("length").is_null())
        .with_columns((pl.col("arrival_time") - pl.col("departure_time")
                      ).alias("travel_time"))
        .sort("trip_id")
    )
    legs = legs.loc[legs["id"].isin(mp_trips["trip_id"])].copy()
    routes = routes.copy()
    legs.sort_values("id", inplace=True)
    
    # Add a join for legs and routes to add the travel_time to legs
    legs = pd.merge(legs, routes, how="left", on=["id", "id"])
    
    legs["trav_time"] = pd.to_timedelta(legs["trav_time"]).dt.total_seconds()
    bins = np.arange(0, m, 0.5)
    fig, ax = get_figure(fraction=0.8)
    ax.hist(
        mp_trips["travel_time"] / 60,
        bins=bins,
        alpha=0.7,
        color=CMP(1),
        label="METROPOLIS2",
        histtype="step",
        density=True,
    )
    ax.hist(
        legs["trav_time"] / 60,
        bins=bins,
        alpha=0.7,
        color=CMP(2),
        label="MATSim",
        histtype="step",
        density=True,
    )
    ax.set_xlabel("Travel time (min.)")
    ax.set_ylabel("Frequency")
    ax.set_xlim(0, m)
    ax.set_ylim(bottom=0)
    ax.legend()
    fig.tight_layout()
    if filename:
        fig.savefig(filename)
    else:
        fig.show()
```

```python
def travel_time_cdf(filename, legs, routes, mp_trips, bin_width=2):
    m = 120  # Max travel time (in minutes)
    mp_trips = (
        mp_trips.with_columns((pl.col("arrival_time") - pl.col("departure_time")
                      ).alias("travel_time"))
        .sort("trip_id")
    )
    legs = legs.loc[legs["id"].isin(mp_trips["trip_id"])].copy()
    routes = routes.copy()
    legs.sort_values("id", inplace=True)

    # Merge to get trav_time
    legs = pd.merge(legs, routes, how="left", on=["id", "id"])
    legs["trav_time"] = pd.to_timedelta(legs["trav_time"]).dt.total_seconds()

    # Convert to minutes
    mp_time = mp_trips["travel_time"] / 60
    matsim_time = legs["trav_time"] / 60

    bins = np.arange(0, m + bin_width, bin_width)

    fig, ax = get_figure(fraction=0.8)

    ax.hist(
        mp_time,
        bins=bins,
        alpha=0.7,
        color=CMP(1),
        label="METROPOLIS2",
        histtype="step",
        density=True,
        cumulative=True,
    )
    ax.hist(
        matsim_time,
        bins=bins,
        alpha=0.7,
        color=CMP(2),
        label="MATSim",
        histtype="step",
        density=True,
        cumulative=True,
    )

    ax.set_xlabel("Travel time (min.)")
    ax.set_ylabel("Cumulative frequency")
    ax.set_xlim(0, m)
    ax.set_ylim(0, 1)
    ax.grid(True, linestyle="--", alpha=0.5)
    ax.legend(fontsize=5)
    fig.tight_layout()

    if filename:
        fig.savefig(filename, dpi=300)
    else:
        fig.show()
```
